```python
import pandas as pd
import numpy as np
```

```python
movies_metadata = pd.read_csv("movies_nonnull.csv")
movies_metadata.head()
```

```python
movies_metadata.info()
```

```python
movies_metadata.drop(columns=['Unnamed: 0','adult','id','original_title','video'], inplace=True)
```

```python
def how_many_not_int(x):
    try:
        return int(x)
    except:
        return np.nan
    
movies_metadata['budget'] = movies_metadata['budget'].apply(how_many_not_int)
movies_metadata['budget'].isna().value_counts()
movies_metadata['budget'].fillna(0)
movies_metadata['budget'] = movies_metadata['budget'].replace(0.0, movies_metadata['budget'].median())
```

```python
import ast
# movies_metadata['content'] = pd.Series([[]]*len(movies_metadata))
# print(movies_metadata['content'])
def extract_from_genre(x):
        t = []
        for j in ast.literal_eval(x):
            try:
                t.append(j['name'])
            except:
                pass
        return t
movies_metadata['genres'] = movies_metadata['genres'].apply(extract_from_genre)
movies_metadata['genres']
```

```python
def clean_lang(x):
    
    try:
        float(x)
        return 'Unknown'
    except:
        return x

movies_metadata['original_language'] = movies_metadata['original_language'].apply(clean_lang)
movies_metadata['original_language'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
# example of a one hot encoding
from numpy import asarray
encoder = OneHotEncoder(sparse=False)
d = asarray(movies_metadata['original_language']).reshape(-1,1)
encoder = OneHotEncoder(sparse=False)
# transform data
onehot = encoder.fit_transform(d)
movies_metadata['original_language'] = pd.Series(onehot.tolist())
```

```python
def extract_from_prod_company(x):
    t = []
    try:
        for j in ast.literal_eval(x):
            t.append(j['name'])
    except:
        pass
    return t
movies_metadata['production_companies'] = movies_metadata['production_companies'].apply(extract_from_prod_company)
movies_metadata['production_companies']
```

```python
movies_metadata = movies_metadata[~movies_metadata['production_companies'].isna()]
```

```python
def extract_from_prod_country(x):
    t = []
    try:
        for j in ast.literal_eval(x):
            t.append(j['name'])
    except:
        pass
    return t
movies_metadata['production_countries'] = movies_metadata['production_countries'].apply(extract_from_prod_country)
movies_metadata['production_countries']
```

```python

def convert_to_datetime(x):
    try:
        return pd.to_datetime(x)
    except:
        return np.nan
movies_metadata['release_date'] = movies_metadata['release_date'].apply(convert_to_datetime)
movies_metadata['release_date'].fillna(movies_metadata['release_date'].mean(), inplace=True)
```

```python
def extract_from_spoken_lan(x):
    t = []
    try:
        for j in ast.literal_eval(x):
            t.append(j['name'])
    except:
        pass
    return t
movies_metadata['spoken_languages'] = movies_metadata['spoken_languages'].apply(extract_from_spoken_lan)
movies_metadata['spoken_languages']
```

```python
def extract_from_tagline(x):
    t = []
    for i in x.split():
        t.append(i)
    
    return t
movies_metadata['tagline'] = movies_metadata['tagline'].apply(extract_from_tagline)
movies_metadata['tagline']
```

```python
movies_metadata['overview']
```

```python
def extract_from_overview(x):
    t = []
    for i in x.split(","):
        for j in i.split():
            t.append(j)
    
    return t
movies_metadata['overview'] = movies_metadata['overview'].apply(extract_from_overview)
movies_metadata['overview']
```

```python
def combine_texts():
    final = []
    for i,j,k in zip(movies_metadata['overview'].values,movies_metadata['tagline'].values, movies_metadata['genres'].values):
        i.extend(j)
        i.extend(k)
    
        final.append(i)
    return final
movies_metadata['keywords'] = pd.Series(combine_texts())
```

```python
movies_metadata['keywords']
```

```python
movies_metadata.drop(columns=['overview','tagline','genres'], inplace=True)
```

```python
movies_metadata.drop(columns=['production_companies','production_countries'], inplace=True)
```

```python
def format_eng_data():
    final = []
    for i in movies_metadata['keywords'].values:
        string = ""
        for j in i:
            if j != '':
                string += j
        final.append(string)
    return final

movies_metadata['keywords'] = pd.Series(format_eng_data())
```

```python
movies_metadata.drop(columns=['original_language'], inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(analyzer='word',stop_words='english', max_features=30000)
X=vectorizer.fit_transform(movies_metadata['keywords'])
tfidf_df=pd.DataFrame.sparse.from_spmatrix(X)
```

```python
movies_df=movies_metadata.join(tfidf_df)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=10,init='k-means++',max_iter=600, verbose=1).fit(movies_df)
```

```python
movies_df['KmeansCluster']=kmeans.labels_

for i in range(10):
    cluster_count = movies_df[movies_df['KmeansCluster']==i].KmeansCluster.count()
    print("Cluster {}: {}".format(i, cluster_count))
```

```python

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim=cosine_similarity(movies_df)
cosine_sim=pd.DataFrame(cosine_sim,index=movies_df.index,columns=movies_df.index)
cosine_sim.head(20)
```
